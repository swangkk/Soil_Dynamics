```python
from scipy import optimize
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
vsarr=[]
vparr=[]
vrarr=[]
parr=[]
p=0
def func(x):
    return x**6-8*x**4+((24-16*a**2)*x**2)+16*(a**2-1)
```

```python
while p < 0.49:
    p+=0.01
    p = round(p,2)
    vs=1
    vp = math.sqrt((2-2*p)/(1-2*p))
    a=1/vp
    sol = optimize.root_scalar(func, bracket=[0, 1], method='brentq')
    vr=sol.root
    vsarr.append(vs)
    vparr.append(vp)
    vrarr.append(vr)
    parr.append(p)
```

```python
# visualization for prediction result
# set figure size
plt.figure(figsize=(10,10))

# set font
font = {'family': 'Arial', 'weight': 'normal', 'size': 16}
plt.rc('font', **font)

# plot data 
plt.plot(parr, vrarr, 'r', label='Vr')
plt.plot(parr, vsarr, 'b', label='Vs')
plt.plot(parr, vparr, 'g', label='Pa')

# set legend position
plt.legend(shadow=True, loc='upper right')

# set x-axis & y-axis titles
#plt.title('prediction result')
plt.xlabel('Poisson Ratio')

# set grid
plt.grid(color='gray', linestyle='--', linewidth=1)

#plt.xticks(0.1)
#plt.yticks(1)
plt.xlim([0, 0.5])
plt.ylim([0, 5])

# save & display figure
plt.show()
```
